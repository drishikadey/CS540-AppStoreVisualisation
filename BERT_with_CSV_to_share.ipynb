```python
#change here  - upload csv into google colab - and change name
#make sure downloads are enabled BEFORE running all 
import pandas as pd
df = pd.read_csv('/content/ios-HealthcareAll.csv',index_col=0)
```

```python
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
#comments =  df["content"].values.tolist()
comments =  df["review"].values.tolist() 
 #change name of column here - should be the name of the column w reviews
```

```python
for i in range(len(comments)):
  if type(comments[i]) !=  str:
    comments[i]=str(comments[i])
  if(len(comments[i])>=512):
    comments[i] = comments[i][:512]
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
returns= sentiment_pipeline(comments)
```

```python
returns
```

```python
df_returns = pd.DataFrame(returns)
df_returns.rename(columns={'label': 'Sentiment BERT Standard', 'score': 'score BERT Standard'}, inplace=True)
```

```python
final = pd.concat([df, df_returns], axis=1)
final
```

```python
sentiment_pipeline2 = pipeline(model = "nlptown/bert-base-multilingual-uncased-sentiment")
returns2 = sentiment_pipeline2(comments)
```

```python
df_returns2 = pd.DataFrame(returns2)
df_returns2.rename(columns={'label': 'Sentiment Review BERT', 'score': 'score Review BERT'}, inplace=True)
```

```python
actual_final = pd.concat([final, df_returns2], axis=1)
```
